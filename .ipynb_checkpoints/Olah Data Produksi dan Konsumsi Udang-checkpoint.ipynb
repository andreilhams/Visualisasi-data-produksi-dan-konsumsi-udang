## Menampilkan data dan visualisasi per-Provinnsi

```python
# Memvisualisasikan data Produksi dan Konsumsi Udang di Provinsi Aceh
fig, ax = plt.subplots(figsize=(16, 10))
ax.plot(df_aceh['Tahun'], df_aceh['Angka Produksi'], 'g--d')
ax.set_title('Angka Produksi Udang di Provinsi Aceh')

ins2 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins2.plot(df_aceh['Tahun'], df_aceh['Angka Konsumsi'])
ins2.set_title('Angka Konsumsi Udang di Provinsi Aceh')

plt.show()
```
